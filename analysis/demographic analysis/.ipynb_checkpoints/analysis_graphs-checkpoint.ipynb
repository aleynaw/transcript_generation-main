```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


# =============================================================================
# Helper Functions
# =============================================================================
def load_data(filepath: str) -> pd.DataFrame:
    """Load the CSV file with a pipe delimiter."""
    return pd.read_csv(filepath, delimiter="|")


def plot_comparison(observed, expected, title, chart_type="bar"):
    """
    Plot a comparison between observed and expected distributions.
    Chart types supported: 'bar' (default) and 'line'.
    """
    categories = list(expected.keys())
    obs_vals = [observed.get(cat, 0) for cat in categories]
    exp_vals = [expected[cat] for cat in categories]
    x = np.arange(len(categories))
    
    fig, ax = plt.subplots(figsize=(10, 6))
    width = 0.35

    if chart_type == "bar":
        rects1 = ax.bar(x - width / 2, exp_vals, width, label='Expected', color='mediumseagreen')
        rects2 = ax.bar(x + width / 2, obs_vals, width, label='Observed', color='dodgerblue')
        for rect in rects1 + rects2:
            height = rect.get_height()
            ax.annotate(f'{height:.1f}',
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),
                        textcoords="offset points",
                        ha='center', va='bottom')
    elif chart_type == "line":
        ax.plot(x, exp_vals, marker='o', linestyle='-', color='mediumseagreen', label='Expected')
        ax.plot(x, obs_vals, marker='o', linestyle='--', color='dodgerblue', label='Observed')
        for i, v in enumerate(exp_vals):
            ax.text(i, v + 1, f"{v:.1f}", ha='center')
        for i, v in enumerate(obs_vals):
            ax.text(i, v + 1, f"{v:.1f}", ha='center')
            
    ax.set_ylabel('Percentage (%)')
    ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels(categories)
    ax.legend()
    plt.tight_layout()
    plt.show()


def get_disabled_category(row):
    """
    Determine the disability category for a patient.
    Checks in order: if not disabled -> 'None'; then Learning, Developmental, Physical.
    If none of the flags are set but the patient is disabled, defaults to 'Physical Disability'.
    """
    if row.get('Disabled', 'No') != "Yes":
        return 'None'
    elif str(row.get('Learning Disability', 'No')).strip() == "Yes":
        return 'Learning Disability'
    elif str(row.get('Developmental Disability', 'No')).strip() == "Yes":
        return 'Developmental Disability'
    elif str(row.get('Physical Disability', 'No')).strip() == "Yes":
        return 'Physical Disability'
    else:
        return 'Physical Disability'


def compute_expected_disability(row):
    """
    Compute the expected disability probabilities based on the patient's age and sex.
    Returns a Series with probabilities (for Learning, Developmental, Physical, and None).
    """
    # Provided probability matrices
    learning_matrix = np.array([[0.085, 0.097],
                                [0.049, 0.056],
                                [0.038, 0.043]])
    developmental_matrix = np.array([[0.033, 0.046],
                                     [0.012, 0.017],
                                     [0.005, 0.007]])
    disabled_base = 0.205  # Overall disabled probability
    non_disabled = 0.795

    sex = row.get("Sex", "Male")
    sex_idx = 0 if sex == "Male" else 1
    age = row.get("Age", 30)
    if age < 25:
        age_idx = 0
    elif age < 65:
        age_idx = 1
    else:
        age_idx = 2

    learning_prob = learning_matrix[age_idx, sex_idx]
    developmental_prob = developmental_matrix[age_idx, sex_idx]
    physical_prob = disabled_base - (learning_prob + developmental_prob)
    return pd.Series({
        "Learning Disability": learning_prob,
        "Developmental Disability": developmental_prob,
        "Physical Disability": physical_prob,
        "None": non_disabled
    })


def assign_age_group(age):
    """
    Assign a simplified age group for certain analyses.
    For education analysis: '18-25', '25-64', '65+'.
    """
    if age < 25:
        return '18-25'
    elif age < 65:
        return '25-64'
    else:
        return '65+'


def assign_age_group_extended(age):
    """
    Assign an extended age group for demographic comparisons.
    Groups: '18-25', '26-35', '36-45', '46-60', '60+'.
    """
    if age < 25:
        return '18-25'
    elif age < 36:
        return '26-35'
    elif age < 46:
        return '36-45'
    elif age < 61:
        return '46-60'
    else:
        return '60+'
```

```python
# =============================================================================
# Main Analysis and Plotting
# =============================================================================
def main():
    # ------------------------------
    # Load Data
    # ------------------------------
    file_path = "llm_patients_1000.csv"
    df = load_data(file_path)

    # ------------------------------
    # Section 1: Feature Comparisons
    # ------------------------------
    # Handedness
    handedness_observed = df['Handedness'].value_counts(normalize=True) * 100
    expected_handedness = {'right': 89, 'left': 10, 'ambidextrous': 1}
    plot_comparison(handedness_observed, expected_handedness, "Handedness Distribution", chart_type="bar")

    # Allergies
    df['Has_Allergies'] = df['Allergies'].apply(
        lambda x: 'Yes' if pd.notnull(x) and str(x).strip() != "" else 'No'
    )
    allergies_observed = df['Has_Allergies'].value_counts(normalize=True) * 100
    expected_allergies = {'Yes': 27.3, 'No': 72.7}
    plot_comparison(allergies_observed, expected_allergies, "Allergies Distribution", chart_type="bar")

    # Concussions (Head Injuries)
    concussion_observed = df['Head Injuries or Concussions'].value_counts(normalize=True) * 100
    expected_concussions = {'Yes': 30, 'No': 70}
    plot_comparison(concussion_observed, expected_concussions, "Concussions (Head Injuries) Distribution", chart_type="bar")

    # Seizures
    seizures_observed = df['History of Seizures'].value_counts(normalize=True) * 100
    expected_seizures = {'Yes': 10, 'No': 90}
    plot_comparison(seizures_observed, expected_seizures, "History of Seizures Distribution", chart_type="bar")

    # Canadian Citizenship
    citizenship_observed = df['Canadian Citizenship'].value_counts(normalize=True) * 100
    expected_citizenship = {'Yes': 91.2, 'No': 8.8}
    plot_comparison(citizenship_observed, expected_citizenship, "Canadian Citizenship Distribution", chart_type="bar")

    # Born in Canada (for Canadian citizens)
    df['Born_in_Canada'] = df['Birthplace'].apply(
        lambda x: 'Yes' if isinstance(x, str) and 'Canada' in x else 'No'
    )
    df_can = df[df['Canadian Citizenship'] == "Yes"]
    born_observed = df_can['Born_in_Canada'].value_counts(normalize=True) * 100
    expected_born = {'Yes': 53.1, 'No': 46.9}
    plot_comparison(born_observed, expected_born, "Born in Canada (for Canadian Citizens) Distribution", chart_type="bar")

    # ------------------------------
    # Section 2: Disability Breakdown
    # ------------------------------
    disability_categories = ["Learning Disability", "Developmental Disability", "Physical Disability", "None"]

    # Determine each patient's disability category
    df['Disabled_Category'] = df.apply(get_disabled_category, axis=1)

    # Observed Disability Breakdown (Pie Chart)
    observed_counts = df['Disabled_Category'].value_counts()
    observed_data = [observed_counts.get(cat, 0) for cat in disability_categories]
    colors = sns.color_palette("pastel", n_colors=len(disability_categories))
    plt.figure(figsize=(6, 6))
    plt.pie(observed_data, labels=disability_categories, autopct='%1.1f%%', startangle=140, colors=colors)
    plt.title('Disability Breakdown (Observed)')
    plt.show()

    # Expected Disability Breakdown (Pie Chart)
    expected_disability_df = df.apply(compute_expected_disability, axis=1)
    expected_disability_avg = expected_disability_df.mean() * 100  # Convert to percentages
    expected_data = [expected_disability_avg.get(cat, 0) for cat in disability_categories]
    plt.figure(figsize=(6, 6))
    plt.pie(expected_data, labels=disability_categories, autopct='%1.1f%%', startangle=140, colors=colors)
    plt.title('Disability Breakdown (Expected)')
    plt.show()

    # ------------------------------
    # Section 3: Future Education Distribution
    # ------------------------------
    # For this analysis, use a simplified age grouping
    df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
    df['Age_Group'] = df['Age'].apply(assign_age_group)
    expected_edu = {
        '18-25': {"Pursuing University Degree": 53, "Pursuing Post-Secondary Certificate or Diploma": 23, "No Further Education": 24},
        '25-64': {"University Degree": 53, "Post-Secondary Certificate or Diploma": 23, "No Further Education": 24},
        '65+': {"University Degree": 28, "Post-Secondary Certificate or Diploma": 22, "No Further Education": 50}
    }
    for group in ['18-25', '25-64', '65+']:
        group_df = df[df['Age_Group'] == group]
        edu_observed = group_df['Further Education After Highschool'].value_counts(normalize=True) * 100
        title = f"Further Education Distribution ({group})"
        plot_comparison(edu_observed, expected_edu[group], title, chart_type="bar")

    # ------------------------------
    # Section 4: Ethnicity Distribution Analysis
    # ------------------------------
    ETHNICITY_CATEGORIES = {
        "Middle Eastern": ["Persian - Iran", "Hebrew - Israel"],
        "European": [
            "Czech - Czech Republic", "Danish - Denmark", "German - Austria", "German - Switzerland", "German - Germany",
            "Greek - Greece", "English - Canada", "English - United States", "English - United Kingdom", "English - Ireland",
            "Spanish - Spain", "Finnish - Finland", "French - Canada", "French - Switzerland", "French - France",
            "Croatian - Croatia", "Hungarian - Hungary", "Armenian - Armenia", "Italian - Italy", "Dutch - Belgium",
            "Dutch - Netherlands", "Norwegian - Norway", "Polish - Poland", "Portuguese - Portugal", "Romanian - Romania",
            "Russian - Russia", "Slovak - Slovakia", "Swedish - Sweden", "Ukrainian - Ukraine"
        ],
        "East Asian": ["Japanese - Japan", "Korean - South Korea", "Chinese - China", "Chinese - Taiwan"],
        "South Asian": ["English - India", "Hindi - India", "Nepali - Nepal", "Tamil - India", "Bengali - Bangladesh",
                        "English - Bangladesh", "Georgian - Georgia", "Azerbaijani - Azerbaijan"],
        "Southeast Asian": ["English - Malaysia", "English - Philippines", "Filipino - Philippines", "Indonesian - Indonesia",
                            "Thai - Thailand", "Tagalog - Philippines", "Vietnamese - Vietnam"],
        "Latin American": ["Spanish - Argentina", "Spanish - Chile", "Spanish - Colombia", "Spanish - Mexico", "Portuguese - Brazil"],
        "African": ["Zulu - South Africa"],
        "Other": ["English - Australia", "English - New Zealand"],
    }

    def categorize_ethnicity(ethnicity):
        for category, values in ETHNICITY_CATEGORIES.items():
            if ethnicity in values:
                return category
        return "Uncategorized"

    df["Ethnicity_Category"] = df["Ethnicity"].apply(categorize_ethnicity)

    # Preprocess additional columns
    df.rename(columns={
        "Sex": "Gender",
        "Relationship Status": "Relationship_Status",
        "Children": "Has_Children",
        "Current Doctor(s)": "Has_Current_Doctor",
        "Canadian Citizenship": "Is_Canadian"
    }, inplace=True)

    # For detailed demographic comparisons, use an extended age grouping
    df['Age_Group'] = df['Age'].apply(assign_age_group_extended)
    relationship_mapping = {
        "Common-Law": "Married/Common-Law",
        "Married": "Married/Common-Law",
        "Divorced": "Divorced/Separated",
        "Separated": "Divorced/Separated"
    }
    df["Relationship_Status"] = df["Relationship_Status"].replace(relationship_mapping)
    df["Gender"] = df["Gender"].astype("category")
    df["Ethnicity_Category"] = df["Ethnicity_Category"].astype("category")
    df["Relationship_Status"] = df["Relationship_Status"].astype("category")
    df["Has_Children"] = df["Has_Children"].apply(lambda x: "Yes" if str(x).strip().lower() == "yes" else "No")
    df["Has_Current_Doctor"] = df["Has_Current_Doctor"].apply(lambda x: "Yes" if pd.notna(x) else "No")
    df["Is_Canadian"] = df["Is_Canadian"].apply(lambda x: "Yes" if str(x).strip().lower() == "yes" else "No")

    # ------------------------------
    # Section 5: Demographic Comparisons
    # ------------------------------
    # Gender Distribution
    plt.figure(figsize=(6, 6))
    ax = sns.countplot(data=df, x="Gender", palette="pastel")
    total = len(df)
    for p in ax.patches:
        ax.annotate(f'{p.get_height()} ({p.get_height() / total * 100:.1f}%)',
                    (p.get_x() + p.get_width() / 2, p.get_height()),
                    ha='center', va='bottom')
    plt.title("Gender Distribution")
    plt.xlabel("Gender")
    plt.ylabel("Count")
    plt.show()

    # Compare Gender Distribution with Actual Demographics
    gender_weights = [0.49, 0.51]
    gender_labels = ["Male", "Female"]
    df_gender_counts = df["Gender"].value_counts(normalize=True)
    actual_gender_dist = pd.Series(gender_weights, index=gender_labels)
    comparison_df = pd.DataFrame({"Dataset": df_gender_counts, "Actual": actual_gender_dist})
    comparison_df.plot(kind="bar", figsize=(6, 6), colormap="Set2")
    plt.title("Comparison of Gender Distribution")
    plt.xlabel("Gender")
    plt.ylabel("Proportion")
    plt.legend(["Dataset", "Actual Demographics"], loc="lower center", ncol=2)
    plt.show()

    # Age Distribution Comparisons
    age_labels_order = ["18-25", "26-35", "36-45", "46-60", "60+"]

    # Pie Charts by Gender
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))
    male_age_counts = df[df["Gender"] == "Male"]["Age_Group"].value_counts().reindex(age_labels_order, fill_value=0)
    axes[0].pie(male_age_counts, labels=age_labels_order,
                autopct=lambda p: f'{p:.1f}%\n({int(p * sum(male_age_counts) / 100)})',
                startangle=90, colors=sns.color_palette("Set2"))
    axes[0].set_title("Age Distribution - Men")
    female_age_counts = df[df["Gender"] == "Female"]["Age_Group"].value_counts().reindex(age_labels_order, fill_value=0)
    axes[1].pie(female_age_counts, labels=age_labels_order,
                autopct=lambda p: f'{p:.1f}%\n({int(p * sum(female_age_counts) / 100)})',
                startangle=90, colors=sns.color_palette("Set2"))
    axes[1].set_title("Age Distribution - Women")
    plt.tight_layout()
    plt.show()

    # Stacked Bar Comparison for Age Distribution (Observed vs. Actual)
    observed_age_counts = df["Age_Group"].value_counts().reindex(age_labels_order, fill_value=0)
    observed_combined = observed_age_counts / observed_age_counts.sum()
    total_sample = observed_age_counts.sum()

    # Actual age distribution (averaged for men and women)
    age_weights_men = [0.2979, 0.1916, 0.1557, 0.2266, 0.1282]
    age_weights_women = [0.2861, 0.1683, 0.1483, 0.2343, 0.1630]
    actual_combined = np.array([(m + w) / 2 for m, w in zip(age_weights_men, age_weights_women)])
    actual_counts = (actual_combined * total_sample).astype(int)

    x = np.arange(len(age_labels_order))
    width = 0.5
    fig, ax = plt.subplots(figsize=(10, 6))
    bars1 = ax.bar(x, actual_combined, width, label="Actual Distribution", color=sns.color_palette("Set2")[0])
    bars2 = ax.bar(x, observed_combined, width, bottom=actual_combined,
                   label="Observed Distribution", color=sns.color_palette("Set2")[1])
    for i, (bar1, bar2) in enumerate(zip(bars1, bars2)):
        height1 = bar1.get_height()
        height2 = bar2.get_height()
        ax.text(bar1.get_x() + bar1.get_width() / 2, height1 / 2,
                f"{actual_combined[i]*100:.1f}%\n({actual_counts[i]})",
                ha='center', va='center', color='black', fontsize=10, fontweight='bold')
        ax.text(bar2.get_x() + bar2.get_width() / 2, height1 + height2 / 2,
                f"{observed_combined[i]*100:.1f}%\n({observed_age_counts.iloc[i]})",
                ha='center', va='center', color='black', fontsize=10, fontweight='bold')
    ax.set_xticks(x)
    ax.set_xticklabels(age_labels_order)
    ax.set_xlabel("Age Group")
    ax.set_ylabel("Proportion")
    ax.set_title("Comparison of Age Distribution (Observed vs. Actual)")
    ax.legend(loc="upper left", bbox_to_anchor=(1, 1))
    plt.tight_layout()
    plt.show()

    # Ethnicity Distribution by Category
    plt.figure(figsize=(10, 5))
    ax = sns.countplot(data=df, y="Ethnicity_Category", palette="pastel",
                       order=df["Ethnicity_Category"].value_counts().index)
    total = len(df)
    for p in ax.patches:
        percentage = f'{100 * p.get_width() / total:.1f}%'
        ax.annotate(percentage, (p.get_width(), p.get_y() + p.get_height() / 2),
                    ha='left', va='center', fontsize=10, color='black', xytext=(5, 0), textcoords='offset points')
    plt.title("Ethnicity Distribution by Category")
    plt.xlabel("Count")
    plt.ylabel("Ethnicity Category")
    plt.show()

    ethnicity_weights = [0.0334, 0.4313, 0.2328, 0.1417, 0.0763, 0.0198, 0.0158, 0.0251]
    ethnicity_labels = list(ETHNICITY_CATEGORIES.keys())
    df_ethnicity_counts = df["Ethnicity_Category"].value_counts(normalize=True)
    actual_ethnicity_dist = pd.Series(ethnicity_weights, index=ethnicity_labels)
    comparison_df = pd.DataFrame({"Dataset": df_ethnicity_counts, "Actual": actual_ethnicity_dist})
    comparison_df.plot(kind="bar", figsize=(10, 5), colormap="tab20")
    plt.title("Comparison of Ethnicity Distribution")
    plt.xlabel("Ethnicity")
    plt.ylabel("Proportion")
    plt.xticks(rotation=45, ha="right")
    plt.legend(["Dataset", "Actual Demographics"])
    plt.show()

    # Relationship Status by Age Group (Heatmap)
    relationship_order = ["Single", "Long-term relationship", "Married/Common-Law", "Divorced/Separated", "Widowded"]
    df['Relationship_Status'] = df['Relationship_Status'].cat.reorder_categories(relationship_order, ordered=True)
    plt.figure(figsize=(8, 5))
    sns.heatmap(pd.crosstab(df["Age_Group"], df["Relationship_Status"], normalize="index") * 100,
                cmap="Blues", annot=True, fmt=".2f")
    plt.title("Relationship Status by Age Group (%)")
    plt.xlabel("Relationship Status")
    plt.ylabel("Age Group")
    plt.show()

    # Actual Demographic Relationship Weights (Heatmap)
    relationship_status_labels = ["Single", "Long-term Relationship", "Married/Common-law", "Divorced/Separated", "Widowed"]
    relationship_weights = [
        [67.35, 25, 7.4, 0.25, 0.0],
        [30.3, 12, 55.35, 2.15, 0.2],
        [7.5, 7, 76.0, 8.15, 1.35],
        [0, 12, 68.33, 15.33, 8.17],
        [0, 15, 54.17, 8.83, 33.5]
    ]
    relationship_weights_df = pd.DataFrame(relationship_weights, index=age_labels_order, columns=relationship_status_labels)
    plt.figure(figsize=(8, 5))
    sns.heatmap(relationship_weights_df, cmap="Oranges", annot=True, fmt=".2f")
    plt.title("Actual Demographic Relationship Status by Age Group (%)")
    plt.xlabel("Relationship Status")
    plt.ylabel("Age Group")
    plt.show()

    # Has Children by Age Group (Heatmap)
    plt.figure(figsize=(8, 5))
    sns.heatmap(pd.crosstab(df["Age_Group"], df["Has_Children"], normalize="index") * 100,
                cmap="Greens", annot=True, fmt=".2f")
    plt.title("Has Children by Age Group (%)")
    plt.xlabel("Has Children")
    plt.ylabel("Age Group")
    plt.show()

    # Actual Demographic Has Children Weights (Heatmap)
    child_probability_matrix = [
        [5, 10, 30, 5, 1],
        [20, 30, 70, 50, 30],
        [40, 50, 85, 70, 60],
        [50, 60, 90, 80, 80],
        [40, 50, 80, 70, 90]
    ]
    actual_child_yes = [np.mean(row) for row in child_probability_matrix]
    actual_child_no = [100 - x for x in actual_child_yes]
    actual_child_df = pd.DataFrame({"No": actual_child_no, "Yes": actual_child_yes}, index=age_labels_order)
    plt.figure(figsize=(8, 5))
    sns.heatmap(actual_child_df, cmap="Oranges", annot=True, fmt=".2f")
    plt.title("Actual Demographic Has Children by Age Group (%)")
    plt.xlabel("Has Children")
    plt.ylabel("Age Group")
    plt.show()

    # Has Current Doctor
    plt.figure(figsize=(6, 4))
    ax = sns.countplot(data=df, x="Has_Current_Doctor", palette="pastel")
    for p in ax.patches:
        ax.annotate(f'{p.get_height()} ({p.get_height() / len(df) * 100:.1f}%)',
                    (p.get_x() + p.get_width() / 2, p.get_height()),
                    ha='center', va='bottom')
    plt.title("Has Current Doctor")
    plt.xlabel("Has Current Doctor")
    plt.ylabel("Count")
    plt.show()

    # Compare Current Doctor with Actual Demographics
    doctor_weights = [0.8, 0.2]
    doctor_labels = ["Yes", "No"]
    df_doctor_counts = df["Has_Current_Doctor"].value_counts(normalize=True).reindex(doctor_labels, fill_value=0)
    actual_doctor_dist = pd.Series(doctor_weights, index=doctor_labels)
    comparison_df = pd.DataFrame({"Dataset": df_doctor_counts, "Actual": actual_doctor_dist})
    comparison_df.plot(kind="bar", figsize=(6, 4), colormap="Set2")
    plt.title("Comparison of Having a Current Doctor")
    plt.xlabel("Has Current Doctor")
    plt.ylabel("Proportion")
    plt.xticks(rotation=0)
    plt.legend(["Dataset", "Actual Demographics"])
    plt.show()

    # Canadian Citizenship
    plt.figure(figsize=(6, 4))
    ax = sns.countplot(data=df, x="Is_Canadian", palette="Set2")
    for p in ax.patches:
        ax.annotate(f'{p.get_height()} ({p.get_height() / len(df) * 100:.1f}%)',
                    (p.get_x() + p.get_width() / 2, p.get_height()),
                    ha='center', va='bottom')
    plt.title("Canadian Citizenship Status")
    plt.xlabel("Is Canadian")
    plt.ylabel("Count")
    plt.show()

    canadian_weights = [0.531, 0.469]
    canadian_labels = ["Yes", "No"]
    df["Is_Canadian"] = pd.Categorical(df["Is_Canadian"].str.strip(), categories=canadian_labels, ordered=True)
    df_canadian_counts = df["Is_Canadian"].value_counts(normalize=True).reindex(canadian_labels, fill_value=0)
    actual_canadian_dist = pd.Series(canadian_weights, index=canadian_labels)
    comparison_df = pd.DataFrame({"Dataset": df_canadian_counts, "Actual": actual_canadian_dist})
    comparison_df.plot(kind="bar", figsize=(6, 4), colormap="Set2")
    plt.title("Comparison of Canadian Citizenship")
    plt.xlabel("Canadian Citizenship")
    plt.ylabel("Proportion")
    plt.xticks(rotation=0)
    plt.legend(["Dataset", "Actual Demographics"])
    plt.show()


if __name__ == "__main__":
    main()
```
